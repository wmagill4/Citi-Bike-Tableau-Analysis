```python
#dependencies
import pandas as pd
```

```python
jan17 = pd.read_csv('Data/201701-citibike-tripdata.csv')
feb17 = pd.read_csv('Data/201702-citibike-tripdata.csv')
mar17 = pd.read_csv('Data/201703-citibike-tripdata.csv')
apr17 = pd.read_csv('Data/201704-citibike-tripdata.csv')
may17 = pd.read_csv('Data/201705-citibike-tripdata.csv')
jun17 = pd.read_csv('Data/201706-citibike-tripdata.csv')
jul17 = pd.read_csv('Data/201707-citibike-tripdata.csv')
aug17 = pd.read_csv('Data/201708-citibike-tripdata.csv')
sep17 = pd.read_csv('Data/201709-citibike-tripdata.csv')
oct17 = pd.read_csv('Data/201710-citibike-tripdata.csv')
nov17 = pd.read_csv('Data/201711-citibike-tripdata.csv')
dec17 = pd.read_csv('Data/201712-citibike-tripdata.csv')
```

```python
jan17.info
```

```python
jan17.head()
```

```python
#Concatenate all 2017 bike station dfs into a single df
bike17 = pd.concat([jan17,feb17,mar17,apr17,may17,jun17,jul17,aug17,sep17,oct17,nov17,dec17], ignore_index=True)
```

```python
bike17.info()
```

```python
#export to csv
bike17.to_csv('Data/2017_citi_bike_trips.csv')
```
